### Features: 
    - id: The id of the movie(clear/unique identifier)
    - title: The official title of the movie.
    - tagline: the tagline of the movie.
    - release_date: Theatrical release date of the movie.
    - genres: Genres associated with the movie 
    - belongs_to_collection: Gives information on the movie series/franchise the particular films belong to.
    - original_language: the language which the movie was originally shot in
    - budget_musd: the budget of the movie in million dollers
    - revenue_musd: The total revenue of the movie in million dollers
    - production_companies: Production companies involed with the making of the movie 
    - production_countries: Contries where the movie was shot/produce in.
    - vote_count: The number of vote by the user, as counted by TMDB
    - vote_average: the avarage rating of the movie 
    - popularity: The popularity score assigned by TMDB
    - runtime: the runtime of the movie in minuts
    - overview: A brif blurb of the movie
    - spoken_languages: spoken languages in the film
    - poster_path: The URL of the poster path
    - cast: (Main) Actros appearing in the movie
    - cast_size: Number of actors appearing in the movie
    - crew_size: Size of the crew (incl. director, excl.actors)
    - director: Name of the directors
